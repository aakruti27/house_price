```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import matplotlib
matplotlib.rcParams["figure.figsize"]=(10,10)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#DATA CLEANING 

df= pd.read_csv('banglore.csv')
df.head()
```

```python
df1=df.dropna()
```

```python
df1.shape
```

```python
df2=df1.drop(['area_type','availability','society'],axis='columns')
df2.head()
```

```python
df2['size'].unique()
```

```python
df2['bedrooms']= df2['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df3=df2.drop(['size'],axis='columns')
df3.head()
```

```python
df.total_sqft.unique()
```

```python
def isflt(i):
    try:
        float(i)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(isflt)].head()
```

```python
def avgsqft(i):
    val=i.split('-')
    if len(val)==2:
        nval=float(val[0])+float(val[1])
        newval=float(nval/2)
        return newval
    try:
        return float(i)
    except:
        return None
```

```python
df4=df3.copy()
df4['total_sqft']= df4['total_sqft'].apply(avgsqft)
```

```python
df4.head(20)
```

```python
#FEATURE ENGG

df5=df4.copy()
df5['price_per_sq']= df5['price']*100000/df5['total_sqft']
df5.head()
```

```python
locnumbers= df5.groupby('location')['location'].agg('count')
locnumless=locnumbers[locnumbers<10]
df5.location=df5.location.apply(lambda x:'other' if x in locnumless else x)
```

```python
df5.head(15)
```

```python
df5.balcony.unique()
```

```python
#OUTLIER DETECTION AND REMOVAL
df5[df5.total_sqft/df5.bedrooms<300].head()
```

```python
df6=df5[~(df5.total_sqft/df5.bedrooms<300)]
df6.shape
```

```python
df6.price_per_sq.describe()
```

```python
#find mean and std dev for each location and then filter out outliers 
def remoutliers(df):
    final_df=pd.DataFrame()
    for k, subdf in df.groupby('location'):
        meanv=np.mean(subdf.price_per_sq)
        stdev=np.std(subdf.price_per_sq)
        newdf=subdf[(subdf.price_per_sq>(meanv-stdev))&(subdf.price_per_sq<(meanv+stdev))]
        final_df=pd.concat([final_df,newdf],ignore_index=True)
    return final_df
df7=remoutliers(df6)
df7.shape
```

```python
df6.bath.unique()
```

```python
#converting location text to number 
pd.get_dummies(df6.location)
```

```python
df7=df6.drop(['price_per_sq'],axis='columns')
df7.head()
```

```python
dumdf=pd.get_dummies(df7.location)
```

```python
df8=pd.concat([df7,dumdf.drop('other',axis='columns')],axis='columns')
df8.head()
```

```python
df9=df8.drop('location',axis='columns')
df9.head()
```

```python
df10=df9.dropna()
X= df10.drop('price',axis='columns')
X.head()
```

```python
y=df10.price
y.head()
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.10,random_state=10)
```

```python
lrmodel=LinearRegression()
lrmodel.fit(X_train,y_train)
lrmodel.score(X_test,y_test)
```
